### File which reads the dow jones index data and plots prices

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('dow_jones_index.data')
data
```

```python
# plot price share with stock_name
# get only stock
stock_name = "XOM"
stock = data[data.stock == stock_name]
plt.plot(stock.date, stock.low)
plt.yscale("linear")
plt.show()
```
